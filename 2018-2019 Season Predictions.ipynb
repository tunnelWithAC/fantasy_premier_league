```python
import os
import pandas as pd
```

```python
file_dir = os.getcwd()  + '/data/FPL18-GW0.csv'
df = pd.read_csv(file_dir)

df['5_game_forecast'] = df['GW1Forecast'] + df['GW2Forecast'] + df['GW3Forecast'] +\
                          df['GW4Forecast'] + df['GW5Forecast']
 
df['Cost'] = df['Cost'] / 1000000
gk_df = df[df['PositionsList'] == 'GLK']
def_df = df[df['PositionsList'] == 'DEF']
mid_df = df[df['PositionsList'] == 'MID']
fwd_df = df[df['PositionsList'] == 'FWD']

df.sort_values(by=['5_game_forecast'], ascending=False).head(70)\
    [['FirstName', 'Surname', 'Team', 'Cost', '5_game_forecast']]
```
